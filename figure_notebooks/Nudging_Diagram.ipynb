```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx],idx
```

```python
sin_go = 0.6*np.sin(np.arange(0,np.pi*2.9,.01))+(np.arange(0,np.pi*2.9,.01)*.2)
```

```python
plt.figure(figsize=(10,4))
sc1 = 0.6
sc2 = 2.9
sc3 = 0.2

sin_go = sc1*np.sin(np.arange(0,np.pi*sc2,.01))+(np.arange(0,np.pi*sc2,.01)*sc3)+1.3

sin_go_2 = sc1*np.sin(np.arange(0,np.pi*sc2,.01))+(np.arange(0,np.pi*sc2,.01)*.04)+1.


plt.plot(np.arange(0,np.pi*sc2,.01),sin_go,color='xkcd:grey',linewidth=3.4,linestyle='--',label='CAM6')
plt.plot(np.arange(0,np.pi*sc2,.01),sin_go_2,color='xkcd:blue grey',linewidth=3.4,label='ERAi')

plt.plot(np.arange(0,np.pi*sc2,.01),0*np.arange(0,np.pi*sc2,.01),color='k', linewidth=2)

plt.plot([0,0],[-4,5],color='k', linewidth=2)
# plt.arrow(2,sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),2)[1]],4,4,width=0.1,color='xkcd:red',alpha=0.2)
plt.arrow(1,sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),1)[1]],0,(sin_go_2[find_nearest(np.arange(0,np.pi*2.9,.01),1)[1]] - sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),1)[1]])+.4,width=0.1,color='xkcd:dark grey',alpha=0.2,label='nudging')
plt.arrow(2,sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),2)[1]],0,(sin_go_2[find_nearest(np.arange(0,np.pi*2.9,.01),2)[1]] - sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),2)[1]])+.48,width=0.1,color='xkcd:dark grey',alpha=0.2)
plt.arrow(3,sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),3)[1]],0,(sin_go_2[find_nearest(np.arange(0,np.pi*2.9,.01),3)[1]] - sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),3)[1]])+.48,width=0.1,color='xkcd:dark grey',alpha=0.2)
plt.arrow(4,sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),4)[1]],0,(sin_go_2[find_nearest(np.arange(0,np.pi*2.9,.01),4)[1]] - sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),4)[1]])+.48,width=0.1,color='xkcd:dark grey',alpha=0.2)
plt.arrow(5,sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),5)[1]],0,(sin_go_2[find_nearest(np.arange(0,np.pi*2.9,.01),5)[1]] - sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),5)[1]])+.48,width=0.1,color='xkcd:dark grey',alpha=0.2)
plt.arrow(6,sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),6)[1]],0,(sin_go_2[find_nearest(np.arange(0,np.pi*2.9,.01),6)[1]] - sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),6)[1]])+.48,width=0.1,color='xkcd:dark grey',alpha=0.2)
plt.arrow(7,sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),7)[1]],0,(sin_go_2[find_nearest(np.arange(0,np.pi*2.9,.01),7)[1]] - sin_go[find_nearest(np.arange(0,np.pi*2.9,.01),7)[1]])+.48,width=0.1,color='xkcd:dark grey',alpha=0.2)

plt.legend(fontsize=17)


plt.ylim([-.1,4])
plt.xlim([-.1,7.5])

plt.xticks([])
plt.yticks([])

plt.savefig(os.getcwd()+'/Figure_output/Nudging_fig.png',bbox_inches='tight', 
               transparent=False,
               pad_inches=0,dpi=400)
plt.show()
# plt.grid(True)
```
